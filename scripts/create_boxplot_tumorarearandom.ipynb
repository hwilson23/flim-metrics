```python
from numpy.random import default_rng

rng = default_rng()
randnras = rng.choice(210,size=150,replace=False)
randpdgfb2 = rng.choice(204,size=150,replace=False)
randpdgfb3 = rng.choice(307,size=150,replace=False)

print(randnras)
print(randpdgfb2)
print(randpdgfb3)
```

```python
import pandas as pd
import numpy as np
from tifffile import tifffile as tiff

with open('G:\\nras_0417_0422_both_photons\\matchednras_tumorarea.txt', 'r') as f1:
    lines1 = f1.readlines()
print(lines1[1:])
filelistnras = []
for files in lines1[1:]:
    filelistnras.append(files)
print(len(filelistnras))

with open('G:\\pdgfb_take2_photons\\matchedpdgfb2_tumorarea.txt', 'r') as f1:
    lines1 = f1.readlines()
print(lines1[1:])
filelistpdgfb2 = []
for files in lines1[1:]:
    filelistpdgfb2.append(files)
print(len(filelistpdgfb2))

with open('G:\\pdgfb_take3_photons\\matchedpdgfb3_tumorarea.txt', 'r') as f1:
    lines1 = f1.readlines()
print(lines1[1:])
filelistpdgfb3 = []
for files in lines1[1:]:
    filelistpdgfb3.append(files)
print(len(filelistpdgfb3))

loc = ['G:\\nras_0417_0422_both_photons\\','G:\\pdgfb_take2_photons\\','G:\\pdgfb_take3_photons\\']
dataout = pd.DataFrame()
for i in randnras:
    currentim = filelistnras[i]
    imname = currentim.split(';')
    #print(imname)
    print(f"{loc}{imname[0]}")
    image = np.array(tiff.imread(f"{loc[0]}{imname[0]}"))
    print({'nras':np.nanmean(image[image>0])})
    d = pd.DataFrame({"photonsmean": [np.nanmean(image[image>0])],'tissue':'nras'})
    dataout = pd.concat([dataout,d],ignore_index=True)

for i in randpdgfb2:
    currentim = filelistpdgfb2[i]
    imname = currentim.split(';')
    #print(imname)
    print(f"{loc}{imname[0]}")
    image = np.array(tiff.imread(f"{loc[1]}{imname[0]}"))
    print({'pdgfb2':np.nanmean(image[image>0])})
    d = pd.DataFrame({"photonsmean": [np.nanmean(image[image>0])],'tissue':'pdgfb2'})
    dataout = pd.concat([dataout,d],ignore_index=True)

for i in randpdgfb2:
    currentim = filelistpdgfb3[i]
    imname = currentim.split(';')
    #print(imname)
    print(f"{loc}{imname[0]}")
    image = np.array(tiff.imread(f"{loc[2]}{imname[0]}"))
    print({'pdgfb3':np.nanmean(image[image>0])})
    d = pd.DataFrame({"photonsmean": [np.nanmean(image[image>0])],'tissue':'pdgfb3'})
    dataout = pd.concat([dataout,d],ignore_index=True)

dataout
```
