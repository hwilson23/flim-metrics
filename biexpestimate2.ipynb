```python
import numpy as np
import matplotlib.pyplot as plt
from multiexp_simulation import simulate_multi_exponential_decay, plot_multi_exponential_decay
```

```python
num_photons = 5
true_lifetimes = [1.1, 0]  # Two lifetime components (in ns)
true_amplitudes = [1, 0]  # Relative amplitudes (70% and 30%)
time_range = (0, 12.5)  # nanoseconds
time_bins = 256
# background_level = (num_photons*0.2)/256 #background counts per bin
background_level = 0
times_multi, counts_multi = simulate_multi_exponential_decay(
    num_photons=num_photons,
    lifetimes=true_lifetimes,
    fractions=true_amplitudes,
    time_range=time_range,
    time_bins=time_bins,
    background=background_level,
)
```

```python
from tiff_export import export_decay_data_as_3d_tiff, export_decay_3d_withnoise
export_decay_data_as_3d_tiff(times_multi, 
                            counts_multi, 
                            fr"G:\MNtissueproject_CLEANED20250716\biexponential_simulations\MULTI_t1_{true_lifetimes[0]}_t2_{true_lifetimes[1]}_data.tif", 
                            box_size=(40, 40)
                            )
```
